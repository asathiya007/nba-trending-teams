```python
import numpy as np
import datetime
```

```python
path = '../data_preparation/model/classifiedNBAData.csv'
df = pd.read_csv(path, index_col=0).dropna()
df = df[df.Team != "warriors"] # dropping because of data issues
df = df[df.Team != "clippers"] # dropping because of data issues
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = df['Date'].map(lambda x: x.date())
```

```python
df
```

```python
line_data = df.groupby(['Team', 'Date'])['RF Positive', 'RF Negative', 'NB Positive', 'NB Negative', 'LR Positive', 'LR Negative'].mean()
```

```python
pd.set_option("display.max_rows", 20)
line_data
```

```python
counts = df
counts['RF Positive'] = np.where((df['RF Positive'] >= .5), 1, df['RF Positive'])
counts['RF Positive'] = np.where((df['RF Positive'] < .5), 0, df['RF Positive'])

counts['RF Negative'] = np.where((df['RF Negative'] >= .5), 1, df['RF Negative'])
counts['RF Negative'] = np.where((df['RF Negative'] < .5), 0, df['RF Negative'])

counts['NB Positive'] = np.where((df['NB Positive'] >= .5), 1, df['NB Positive'])
counts['NB Positive'] = np.where((df['NB Positive'] < .5), 0, df['NB Positive'])

counts['NB Negative'] = np.where((df['NB Negative'] >= .5), 1, df['NB Negative'])
counts['NB Negative'] = np.where((df['NB Negative'] < .5), 0, df['NB Negative'])

counts['LR Positive'] = np.where((df['LR Positive'] >= .5), 1, df['LR Positive'])
counts['LR Positive'] = np.where((df['LR Positive'] < .5), 0, df['LR Positive'])

counts['LR Negative'] = np.where((df['LR Negative'] >= .5), 1, df['LR Negative'])
counts['LR Negative'] = np.where((df['LR Negative'] < .5), 0, df['LR Negative'])

counts

for_tweets = counts.copy()
```

```python
counts = df.groupby(['Team', 'Date'])['RF Positive', 'RF Negative', 'NB Positive', 'NB Negative', 'LR Positive', 'LR Negative'].sum()
```

```python
counts = counts.astype({'RF Positive': 'int32', 'RF Negative': 'int32', 'NB Positive': 'int32', 'NB Negative': 'int32', 'LR Positive': 'int32', 'LR Negative': 'int32'})
counts
```

```python
print("Range of num tweets: ", counts['RF Positive'].min(),"-", counts['RF Positive'].max())
```
